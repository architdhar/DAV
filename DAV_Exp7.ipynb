### **Tokenization**

```python
from nltk.tokenize import sent_tokenize
text="""Tokenization is the process of breaking down text into words, phrases, symbols, or other meaningful elements, often referred to as tokens
"""

tokenized_text=sent_tokenize(text)
print(tokenized_text)
```

```python
from nltk.tokenize import word_tokenize

tokenized_word=word_tokenize(text)

print(tokenized_word)
```

### **Frequency Distribution**

```python
from nltk.probability import FreqDist

fdist = FreqDist(tokenized_word)
print(fdist.most_common(2))
```

```python
import matplotlib.pyplot as plt

fdist.plot(30,cumulative=False)
plt.show()
```

### **Stopwords**

```python
from nltk.corpus import stopwords

stop_words=set(stopwords.words("english"))

print(stop_words)
```

**Removing Stopwords**

```python
filtered_tokens=[]
for w in tokenized_word:
    if w not in stop_words:
         filtered_tokens.append(w)

print("Tokenized Words:",tokenized_word)
print("Filterd Tokens:",filtered_tokens)
```

**Removing Punctuations**

```python
import string

# punctuations
punctuations=list(string.punctuation)
punctuations.append('â€™')

filtered_tokens2=[]

for i in filtered_tokens:
    if i not in punctuations:
        filtered_tokens2.append(i)

print("Filterd Tokens After Removing Punctuations:",filtered_tokens2)
```

### **Lexicon Normalization**

**Stemming**

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
ps = PorterStemmer()

stemmed_words=[]

for w in filtered_tokens2:
     stemmed_words.append(ps.stem(w))

print("Filtered Tokens After Removing Punctuations:",filtered_tokens2)
print("Stemmed Tokens:",stemmed_words)
```

**Lemmatization**

```python
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.porter import PorterStemmer

lem = WordNetLemmatizer()
stem = PorterStemmer()
word = "flying"

print("Lemmatized Word:", lem.lemmatize(word, "v"))
print("Stemmed Word:", stem.stem(word))

filtered_tokens2 = ["cats", "running", "played"]  # Assuming you have some tokens here

lem_words = []
for w in filtered_tokens2:
    lem_words.append(lem.lemmatize(w))

print("Lemmatized Words:", lem_words)
```

### **POS tagging**

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag

sent = "Albert Einstein was born in Ulm, Germany in 1879."

tokens=word_tokenize(sent)
pos_=pos_tag(tokens)


print("Tokens:",tokens)
print("PoS tags:",pos_)
```

### **Named Entity Recognization**

```python
from nltk import ne_chunk

sent="New York City on Tuesday declared a public health emergency and ordered mandatory measles vaccinations amid an outbreak, becoming the latest national flash point over refusals to inoculate against dangerous diseases."

for chunk in ne_chunk(nltk.pos_tag(word_tokenize(sent))):
        if hasattr(chunk, 'label'):
           print(chunk.label(), ' '.join(c[0] for c in chunk))
```

### **Web Scraping**

```python
import pandas as pd
import os
import requests
from bs4 import BeautifulSoup
```

```python
def scrape(url):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        title = soup.find('h1').get_text()
        print("Article Title: \n" + title)
        text = soup.find('div', class_= 'td-post-content')

        if text:
          pre_tag = text.find('pre')
          if pre_tag:
            pre_tag.decompose()

        new_text = text.get_text()
        print("\n\nArticle Text: " + new_text)
```

```python
scrape("https://insights.blackcoffer.com/credit-scoring-model/")
```
